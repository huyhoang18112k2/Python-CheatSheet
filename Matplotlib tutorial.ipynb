## Getting data from the internet

```python
import matplotlib.dates as mdates
import urllib

def bytespdate2num(): # String date to number
    def bytesconverter(b):
        s = b.decode('utf-8')
        return (mdates.datestr2num(s))
    return bytesconverter

def graph_data():
    
    stock_prices_url = "https://pythonprogramming.net/yahoo_finance_replacement"
    source_code = urllib.request.urlopen(stock_prices_url).read().decode() #Return string file
    
    stock_data = []
    split_source = source_code.split('\n') # Change to list of string
    
    for line in split_source:
        split_line = line.split(',')
        if len(split_line) == 7:
            if 'Volume' not in line and 'labels' not in line:
                stock_data.append(line)
                
    
    date, openp, highp, lowp, closep, adjustedp, Volume = np.loadtxt(stock_data,
                                                         delimiter=',',
                                                         unpack=True,
                                                         converters={0:bytespdate2num()})
                
    plt.plot_date(date,closep,'-', label = 'Prices BTC')
    plt.xlabel('Dates')
    plt.ylabel('Prices')
    plt.title('BTC prices')
    plt.legend()
    plt.show()
    
graph_data()
```

## Basic customizations, Rotating labels

```python
def bytespdate2num(): # String date to number
    def bytesconverter(b):
        s = b.decode('utf-8')
        return (mdates.datestr2num(s))
    return bytesconverter

def graph_data():
    
    fig = plt.figure()
    ax1 = plt.subplot2grid((1,1), (0,0))
    
    stock_prices_url = "https://pythonprogramming.net/yahoo_finance_replacement"
    source_code = urllib.request.urlopen(stock_prices_url).read().decode() #Return string file
    
    stock_data = []
    split_source = source_code.split('\n') # Change to list of string
    
    for line in split_source:
        split_line = line.split(',')
        if len(split_line) == 7:
            if 'Volume' not in line and 'labels' not in line:
                stock_data.append(line)
                
    
    date, openp, highp, lowp, closep, adjustedp, Volume = np.loadtxt(stock_data,
                                                         delimiter=',',
                                                         unpack=True,
                                                         converters={0:bytespdate2num()})
    #Everything in matplotlib is customizable         
    ax1.plot_date(date,closep,'-', label = 'Prices BTC')
    for label in ax1.xaxis.get_ticklabels():
        label.set_rotation(45)
    
    ax1.grid(True, color = 'r', linestyle = '-', linewidth = 0.5)# (optional)
    
    
    plt.xlabel('Dates')
    plt.ylabel('Prices')
    plt.title('BTC prices')
    plt.subplots_adjust(left=0.09, bottom=0.18, right=0.94, top = 0.95, wspace=0.2, hspace=0)
    plt.legend()
    plt.show()
    
graph_data()
```

##  Candlestick OHLC Graphs

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as mticker
from mpl_finance import candlestick_ohlc

import numpy as np
import urllib
import datetime as dt
```

```python
def bytespdate2num(): # String date to number
    def bytesconverter(b):
        s = b.decode('utf-8')
        return (mdates.datestr2num(s))
    return bytesconverter

def graph_data():
    
    fig = plt.figure()
    ax1 = plt.subplot2grid((1,1), (0,0))
    
    stock_prices_url = "https://pythonprogramming.net/yahoo_finance_replacement"
    source_code = urllib.request.urlopen(stock_prices_url).read().decode() #Return string file
    
    stock_data = []
    split_source = source_code.split('\n') # Change to list of string
    
    for line in split_source:
        split_line = line.split(',')
        if len(split_line) == 7:
            if 'Volume' not in line and 'labels' not in line:
                if len(stock_data)< 100:
                    stock_data.append(line)
                else:
                    break
                            
                        
    date, openp, highp, lowp, closep, adjustedp, Volume = np.loadtxt(stock_data,
                                                         delimiter=',',
                                                         unpack=True,
                                                         converters={0:bytespdate2num()})
    x = 0
    y = len(date)
    ohlc = []
    while(x<y):
        append_me = date[x], openp[x], highp[x],lowp[x],closep[x],Volume[x]
        ohlc.append(append_me)
        x += 1
        
    plt.plot([],[], label = 'BTC prices', color = '#53d06a', linewidth = 4)
    #draw candle OHLC:
    candlestick_ohlc(ax1,ohlc,width = 0.4, colorup = '#53d06a', colordown='#ea2128')
    
    for label in ax1.xaxis.get_ticklabels():
        label.set_rotation(45)
    
    ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    ax1.xaxis.set_major_locator(mticker.MaxNLocator(10))
    ax1.grid(True)
    
    ##Show------------------------------------------------------------------------------------
    plt.xlabel('Dates')
    plt.ylabel('Prices')
    plt.title('BTC prices')
    plt.subplots_adjust(left=0.09, bottom=0.18, right=0.94, top = 0.95, wspace=0.2, hspace=0)
    plt.legend()
    plt.show()
     ##----------------------------------------------------------------------------------------
    
graph_data()
```

## Live Graphs

```python
import matplotlib.animation as animation
from matplotlib import style

style.use('seaborn')

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

def animate(i):
    graph_data = open('example.txt', 'r').read()
    lines = graph_data.split('\n')
    xs = []
    ys = []
    for line in lines:
        if len(line)>1:
            x, y = line.split(',')
            xs.append(x)
            ys.append(y)
    ax1.clear()
    ax1.plot(xs,ys)

ani = animation.FuncAnimation(fig, animate, interval = 1000)
plt.show()
        
```
